```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option('display.max_columns', None)
pd.pandas.set_option('display.max_rows', None)
```

```python
path = '/data/'

X_train = pd.read_csv(path+'xtrain.csv')
X_test = pd.read_csv(path+'xtest.csv')
X_train.head()
```

```python
#loading target (target is log transformed)
y_train = pd.read_csv(path+'ytrain.csv')
y_test = pd.read_csv(path+'ytest.csv')

y_train.head()
```

Let's selec a subset of most predictive features

* I will do model fitting and feature selection altogether in a few lines of code

1. Specify Lasso Regression model and I will
2. Select a suitable alpha (equivalent of penalty)
3. Bigger the alpha, less features that will be selected
4. Now I will use `selectFromModel object` from sklearn
5. I will select automatically features which coefficients are non-zero

```python
sel_ = SelectFromModel(Lasso(alpha=0.001,random_state=108))

#train Lasso model and select features
sel_.fit(X_train,y_train)
```

```python
sel_.get_support().sum()
```

```python
#selected features marked as True

sel_.get_support()
```

```python
# let's print number of total and selected features

# this is how I can make a list of selected features
selected_feats = X_train.columns[(sel_.get_support())]

#some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feats)))
print('features with coefficients shrank to zero: {}'.format(np.sum(sel_.estimator_.coef_==0)))
```
